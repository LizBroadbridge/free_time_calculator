```python
df = pd.read_csv("./split_week_csv/baseline.csv")
df = df[df["Text"].str.startswith("Free")].reset_index(drop=True)
df["Start"] = pd.to_datetime(df["Start"], format="%H:%M")
df["End"] = pd.to_datetime(df["End"], format="%H:%M")
df["Duration"] = (df["End"] - df["Start"]).dt.total_seconds() / 60 / 60
df = df.groupby(["Week", "Text"])["Duration"].sum().reset_index()
multipliers = [4,2]
def multiply_duration(row):
    return multipliers[row['Week']] * row['Duration']
def average_duration(row):
    return round(row['Duration'] / sum(multipliers), 1)
df['Duration'] = df.apply(multiply_duration, axis=1)
df = df.groupby("Text")["Duration"].sum().reset_index()
df['Duration'] = df.apply(average_duration, axis=1)
print(df.to_markdown(index=False))
```
